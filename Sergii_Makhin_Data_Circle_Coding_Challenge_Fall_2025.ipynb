## ðŸ› ï¸ Step 2: Clean the Data

Tasks:
- Handle missing data (replace missings)
- Extract or clean salary values (make sure salary is numerical)

```python
# your code

# Column "Unnamed: 0" seems to be a leftover index column and therefore is redundant. Let's drop it
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
import numpy as np

# Replace -1, -1.0, and '-1' with NaN
df.replace([-1, -1.0, '-1'], np.nan, inplace=True)
```

```python
# Remove columns with extremely high number of missing values
df.drop(columns=['Easy Apply', 'Competitors'], inplace=True)
```

```python
# Remove rows where the number of missing values exceeds 5
df = df[df.isnull().sum(axis=1) <= 5]
df.shape
```

```python
# Check for missing values again
df.isnull().sum()
```

```python
# Check 'Salary Estimate'
print(f"Number of unique values in 'Salary Estimate' equals to {df['Salary Estimate'].nunique()}")
print("A few examples of salary estimates:")
df['Salary Estimate'].value_counts().head(3)
```

```python
# Impute the mode for the only missing value
df['Salary Estimate'] = df['Salary Estimate'].fillna(df['Salary Estimate'].mode()[0])
```

```python
# Let's extract the minimal and top salary estimate for each job listing
# Define function to extract min and max salaries
def extract_salaries(salary_str):
    if pd.isna(salary_str) or not isinstance(salary_str, str):
        return pd.NA, pd.NA
    # Remove the '(Glassdoor est.)' part and split by '-'
    salary_range = salary_str.replace(' (Glassdoor est.)', '').split('-')
    if len(salary_range) == 2:
        min_salary = int(salary_range[0].replace('$', '').replace('K', '000'))
        max_salary = int(salary_range[1].replace('$', '').replace('K', '000'))
        return min_salary, max_salary
    return pd.NA, pd.NA

# Apply the function to create new columns
df[['Min_Salary', 'Max_Salary']] = df['Salary Estimate'].apply(lambda x: pd.Series(extract_salaries(x)))

# Display the result
print(df[['Salary Estimate', 'Min_Salary', 'Max_Salary']].head())
```

```python
# Check 'Rating' column
sns.histplot(df['Rating'])
```

```python
# Looks like we can just impute the mean for the missing values
df['Rating'] = df['Rating'].fillna(df['Rating'].mean())
```

```python
# Check column 'Headquarters'
df['Headquarters'].value_counts().head(5)
```

```python
# Imputing the mode seems a reasonably good idea for 9 missing values
df['Headquarters'] = df['Headquarters'].fillna(df['Headquarters'].mode()[0])
```

```python
# Check the most frequent values in 'Industry'
df['Industry'].value_counts().head(5)
```

```python
# Check the most frequent values in 'Sector'
df['Sector'].value_counts().head(5)
```

```python
# The number of missing values for both columns equals 190 (about 10% of the reamaining dataset).
# The features appear to be essential for further analysis. Imputing the mode would introduce a noticable bias in the data.
# That's why imputing a placeholder like 'Unknown' seems reasonable.

df['Industry'] = df['Industry'].fillna('Unknown')
df['Sector'] = df['Sector'].fillna('Unknown')
```

```python
# Column 'Founded' contains about 25% missing values
# Let's first check the values distibution
sns.histplot(df['Founded'])
```

```python
# Seems reasonable to make categorical bins to group data for further comparison and impute 'Unknown' for missing values

# Define the bins and labels
bins = [0, 1900, 1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
labels = ['Before 1900', '1901-1950', '1951-1960', '1961-1970', '1971-1980', '1981-1990', '1991-2000', '2001-2010', '2010+']

# Verify lengths
print(f"Number of bins: {len(bins)}")
print(f"Number of labels: {len(labels)}")

# Create the new categorical column, handling NaN values
df['Founded_Category'] = pd.cut(df['Founded'], bins=bins, labels=labels, include_lowest=True)
df['Founded_Category'] = df['Founded_Category'].cat.add_categories('Unknown').fillna('Unknown')

# Display the first few rows to verify
print(df[['Founded', 'Founded_Category']].head())
```

```python
# Drop 'Founded' column
df.drop('Founded', axis=1, inplace=True)
```

```python
# Check for missing values
df.isnull().sum()
```

## ðŸ“Š Step 3: Exploratory Data Analysis (EDA)

Explore the job market:
- Top job titles
- Most hiring companies
- Popular locations
- Salary distribution

```python
# Check column 'Job Titles'
print(f"The number of unique values in 'Job Title' column equals to {df['Job Title'].nunique()}")
```

```python
# Check the titles for patterns of characteristic words
print(df['Job Title'].value_counts().head(20))
```

```python
# Let's group titles thematically

# Define function to group job titles
def group_job_title(title):
    title = title.lower()
    if 'data analyst' in title or 'analyst' in title:
        if 'senior' in title or 'sr' in title or 'sr.' in title:
            return 'Senior Data Analyst'
        elif 'junior' in title or 'jr' in title or 'jr.' in title:
            return 'Junior Data Analyst'
        elif 'lead' in title or 'reporting' in title:
            return 'Lead/Reporting Analyst'
        elif 'quality' in title or 'governance' in title:
            return 'Data Quality/Governance Analyst'
        elif 'financial' in title:
            return 'Financial Data Analyst'
        elif 'warehouse' in title or 'management' in title:
            return 'Data Warehouse/Management Analyst'
        elif 'marketing' in title or 'technical' in title or 'sql' in title:
            return 'Specialized Data Analyst'
        else:
            return 'Data Analyst'
    return 'Other'

# Apply grouping
df['Job Title Grouped'] = df['Job Title'].apply(group_job_title)

# Drop the old column as redundant
df = df.drop('Job Title', axis=1)

# Verify the result
print(df['Job Title Grouped'].value_counts())
```

```python
# Visualize the distribution of job listings

job_counts = df['Job Title Grouped'].value_counts().reset_index()
job_counts.columns = ['Job Title', 'Count']

plt.figure(figsize=(9, 5))
sns.barplot(
    data=job_counts,
    y='Job Title',
    x='Count',
    hue='Job Title',
    palette='viridis',
)

plt.xlabel("Number of Job Listings")
plt.ylabel("Job Title")
plt.title("Distribution of Data Analyst Job Titles")
plt.tight_layout()
plt.show()
```

```python
# Check unique values in "Company Name"

df['Company Name'].nunique()
```

```python
print(df['Company Name'].value_counts().head(20))
```

```python
# Remove newline characters and trailing numbers
import re
def clean_company_name(name):
    # Remove \n and trailing version-like numbers (e.g., \n3.5)
    name = re.sub(r'\n\d+(\.\d+)?$', '', name)
    name = name.replace('\n', ' ')  # Replace other \n with space if present
    return name.strip()

# Apply cleaning
df['Company Name'] = df['Company Name'].apply(clean_company_name)
```

```python
#Check the result

print(f"Number of unique values is {df['Company Name'].nunique()}\n")
print(df['Company Name'].value_counts().head(20))
```

```python
# Get top 20 companies by job count
top_companies = df['Company Name'].value_counts().head(20).reset_index()
top_companies.columns = ['Company Name', 'Job Count']

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_companies,
    y='Company Name',
    x='Job Count',
    hue='Company Name',
    palette='mako'
)

plt.title("Top 20 Companies by Number of Job Listings")
plt.xlabel("Number of Listings")
plt.ylabel("Company Name")
plt.tight_layout()
plt.show()
```

```python
# Check column 'Location'
print(f"Number of unique values is {df['Location'].nunique()}\n")
print(df['Location'].value_counts().head(20))
```

```python
# Get top 20 locations by job listings
top_locations = df['Location'].value_counts().head(20).reset_index()
top_locations.columns = ['Location', 'Job Count']

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_locations,
    y='Location',
    x='Job Count',
    hue='Location',
    palette='rocket'
)

plt.title("Top 20 Locations by Number of Job Listings")
plt.xlabel("Number of Listings")
plt.ylabel("Location")
plt.tight_layout()
plt.show()
```

```python
# your code

# Reshape data for violin plot
df_melted = df[['Job Title Grouped', 'Min_Salary', 'Max_Salary']].melt(
    id_vars=['Job Title Grouped'],
    value_vars=['Min_Salary', 'Max_Salary'],
    var_name='Salary_Type',
    value_name='Salary')

plt.figure(figsize=(14, 8))
sns.violinplot(data=df_melted, x='Job Title Grouped', y='Salary', hue='Salary_Type',
               palette='pastel', inner='box', density_norm='width')
plt.title('Salary Distribution by Job Title', fontsize=16, fontweight='bold')
plt.xlabel('Job Title', fontsize=12)
plt.ylabel('Salary ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.ylim(0, df_melted['Salary'].max() * 1.1)  # Set y-axis to start from 0
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Extract the USA state abbreviation from "Location" column
df['State'] = df['Location'].apply(lambda x: x.split(', ')[-1] if ', ' in str(x) else None)
```

```python
# Group by state and calculate salary statistics
salary_by_state = df.groupby('State').agg({
    'Min_Salary': ['mean', 'median', 'min', 'max', 'count'],
    'Max_Salary': ['mean', 'median', 'min', 'max', 'count']
}).round(0)

# Flatten column names
salary_by_state.columns = ['_'.join(col).strip() for col in salary_by_state.columns]

# Calculate average salary range
salary_by_state['Avg_Min_Salary'] = salary_by_state['Min_Salary_mean']
salary_by_state['Avg_Max_Salary'] = salary_by_state['Max_Salary_mean']
salary_by_state['Avg_Salary_Range'] = salary_by_state['Avg_Max_Salary'] - salary_by_state['Avg_Min_Salary']

# Sort by average max salary
salary_by_state = salary_by_state.sort_values('Avg_Max_Salary', ascending=False)

print("Salary Statistics by State:")
print(salary_by_state[['Min_Salary_count', 'Avg_Min_Salary', 'Avg_Max_Salary', 'Avg_Salary_Range']].head(10))

# Create visualizations
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Salary Analysis by State', fontsize=16, fontweight='bold')

# 1. Average Min and Max Salary by State (Top 10)
top_10_states = salary_by_state.head(10)
x_pos = np.arange(len(top_10_states))

axes[0, 0].bar(x_pos - 0.2, top_10_states['Avg_Min_Salary'], 0.4,
               label='Avg Min Salary', color='lightblue', alpha=0.8)
axes[0, 0].bar(x_pos + 0.2, top_10_states['Avg_Max_Salary'], 0.4,
               label='Avg Max Salary', color='darkblue', alpha=0.8)
axes[0, 0].set_xlabel('State')
axes[0, 0].set_ylabel('Salary ($)')
axes[0, 0].set_title('Average Min vs Max Salary by State (Top 10)')
axes[0, 0].set_xticks(x_pos)
axes[0, 0].set_xticklabels(top_10_states.index, rotation=45)
axes[0, 0].legend()
axes[0, 0].grid(axis='y', alpha=0.3)

# Format y-axis to show salary in thousands
axes[0, 0].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))

# 2. Salary Range by State
axes[0, 1].bar(range(len(top_10_states)), top_10_states['Avg_Salary_Range'],
               color='green', alpha=0.7)
axes[0, 1].set_xlabel('State')
axes[0, 1].set_ylabel('Salary Range ($)')
axes[0, 1].set_title('Average Salary Range by State (Top 10)')
axes[0, 1].set_xticks(range(len(top_10_states)))
axes[0, 1].set_xticklabels(top_10_states.index, rotation=45)
axes[0, 1].grid(axis='y', alpha=0.3)
axes[0, 1].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))

# 3. Number of Job Listings by State
job_counts = salary_by_state['Min_Salary_count'].sort_values(ascending=False).head(10)
axes[1, 0].bar(range(len(job_counts)), job_counts.values, color='orange', alpha=0.7)
axes[1, 0].set_xlabel('State')
axes[1, 0].set_ylabel('Number of Job Listings')
axes[1, 0].set_title('Number of Job Listings by State (Top 10)')
axes[1, 0].set_xticks(range(len(job_counts)))
axes[1, 0].set_xticklabels(job_counts.index, rotation=45)
axes[1, 0].grid(axis='y', alpha=0.3)

# 4. Box plot of salary distributions for top states
top_5_states = salary_by_state.head(5).index
df_top_states = df[df['State'].isin(top_5_states)]

# Prepare data for box plot
salary_data = []
state_labels = []
for state in top_5_states:
    state_data = df_top_states[df_top_states['State'] == state]
    salary_data.extend(state_data['Min_Salary'].tolist())
    salary_data.extend(state_data['Max_Salary'].tolist())
    state_labels.extend([f'{state}_Min'] * len(state_data))
    state_labels.extend([f'{state}_Max'] * len(state_data))

# Create box plot data
box_df = pd.DataFrame({'Salary': salary_data, 'State_Type': state_labels})
sns.boxplot(data=box_df, x='State_Type', y='Salary', ax=axes[1, 1])
axes[1, 1].set_xlabel('State and Salary Type')
axes[1, 1].set_ylabel('Salary ($)')
axes[1, 1].set_title('Salary Distribution by State (Top 5)')
axes[1, 1].tick_params(axis='x', rotation=45)
axes[1, 1].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))

plt.tight_layout()
plt.show()

# Additional summary table
print("\n" + "="*50)
print("SUMMARY TABLE - Top 10 States by Average Max Salary")
print("="*50)
summary_table = salary_by_state[['Min_Salary_count', 'Avg_Min_Salary', 'Avg_Max_Salary', 'Avg_Salary_Range']].head(10)
summary_table.columns = ['Job_Count', 'Avg_Min_Salary', 'Avg_Max_Salary', 'Salary_Range']
print(summary_table.to_string())

# Create a separate detailed comparison chart
plt.figure(figsize=(14, 8))
states_to_plot = salary_by_state.head(12).index
x_pos = np.arange(len(states_to_plot))

plt.bar(x_pos - 0.2, salary_by_state.loc[states_to_plot, 'Avg_Min_Salary'], 0.4,
        label='Average Min Salary', color='lightcoral', alpha=0.8)
plt.bar(x_pos + 0.2, salary_by_state.loc[states_to_plot, 'Avg_Max_Salary'], 0.4,
        label='Average Max Salary', color='steelblue', alpha=0.8)

plt.xlabel('State', fontsize=12)
plt.ylabel('Salary ($)', fontsize=12)
plt.title('Average Salary Comparison by State (Top 12)', fontsize=14, fontweight='bold')
plt.xticks(x_pos, states_to_plot, rotation=45, ha='right')
plt.legend(fontsize=11)
plt.grid(axis='y', alpha=0.3)
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))

# Add value labels on bars
for i, state in enumerate(states_to_plot):
    min_val = salary_by_state.loc[state, 'Avg_Min_Salary']
    max_val = salary_by_state.loc[state, 'Avg_Max_Salary']
    plt.text(i - 0.2, min_val + 2000, f'${min_val/1000:.0f}K', ha='center', va='bottom', fontsize=9)
    plt.text(i + 0.2, max_val + 2000, f'${max_val/1000:.0f}K', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

## We have observed a few patterns:

**Two Clear Groups**: Salaries fall into two separate buckets - starting salaries cluster around 45-50K USD. Top salaries spread widely from 80K-200K USD with little overlap between them.

**Experience Matters**: Entry-level jobs pay pretty much the same everywhere, but senior roles vary wildly - some pay 80K USD, others 200K USD for similar titles.

**Predictable vs Unpredictable**: New job salaries are very predictable, but experienced role salaries depend heavily on the specific company, location, and your skills.

## What should a job seeker consider?

**Starting Out**: Expect around $45-50K as a new data analyst - this is pretty standard across the board, so don't stress too much about negotiating your first salary.

**Plan Your Growth**: The big money is in senior roles, so focus on building specialized skills early. Your salary could double or triple with the right experience and expertise.

**Do Your Homework**: For senior positions, research is crucial. The same job title at different companies could pay $80K or $180K - company size, industry, and location make huge differences.

**Specialize Early**: "Specialized Data Analyst" roles often pay the most. Pick a niche (like machine learning or healthcare analytics) rather than staying too general.

## Surprising insights:

**No Middle Ground**: There's a clear jump from "junior" to "senior" pay - no gradual increases. You're either in the 45-50K USD tier or the 80K+ USD tier, with little in between.

**Senior Roles Are Chaotic**: Experienced analyst salaries are all over the map. The market seems less organized at senior levels, creating both opportunities and risks.

**Entry-Level Cap**: Starting salaries seem to hit a ceiling around 60K USD, even for exceptional candidates. The real earning potential comes later in your career.
